```python
import requests
import json
from APIKEY import eia_api
```

```python
### Energy Production ###
# Coal, Crude Oil, Geothermal, Hydroelectrical, Natural Gas (Dry),Natural Gas (Liquid) Nuclear, Solar/PV, Wind
energy_sources = ['TOTAL.CLPRBUS.A','TOTAL.PAPRBUS.A','TOTAL.GETCBUS.A','TOTAL.HVTCBUS.A','TOTAL.NGPRBUS.A','TOTAL.NLPRBUS.A','TOTAL.NUETBUS.A','TOTAL.SOTCBUS.A','TOTAL.WYTCBUS.A']

url = "http://api.eia.gov/series/?api_key="
series = '&series_id='

for source in energy_sources:
    query_url = url + eia_api + series + source
    print(query_url)
```

```python
request = requests.get(query_url)
request
```
