```python
# Dependencies
from splinter import Browser
from bs4 import BeautifulSoup
import pandas as pd
import re
```

```python
# Setting up chromedriver
executable_path = {'executable_path': '/usr/local/bin/chromedriver'}
browser = Browser('chrome', **executable_path, headless=False)
```

```python
url = 'https://www.indeed.com/jobs?q=data+analyst&l=United+States'
browser.visit(url)
```

```python
counter = 0
html = browser.html
soup = BeautifulSoup(html, 'html.parser')
total_pages_string = soup.find('div', id='searchCountPages').text
total_pages_string = total_pages_string.split()
total_pages = int(total_pages_string[3].replace(",", ""))
print(total_pages)

for x in range(1, 10):
    
    html = browser.html
    soup = BeautifulSoup(html, 'html.parser')
    print(f'Page: {x}')
    print('-'*80)

    postings = soup.find_all('div', class_='jobsearch-SerpJobCard')

    for posting in postings:
        counter += 1
        
        # Get company name
        try:
            company_name_tag = posting.find('span',class_="company")
            company_name = company_name_tag.find('a').text
        except:
            company_name = posting.find('span',class_="company").text
            
        # Get job title
        job_title = posting.find('h2', class_="title").text

        # Get location
        try:
            job_location = posting.find('div', class_="location").text
        except:
            job_location = posting.find('span', class_="location").text

        # Get salary
        try:
            salary = posting.find('span', class_="salaryText").text
        except:
            salary = "NA"
            
        print(f'Posting #{counter}')  
        print(company_name)
        print(job_title)
        print(job_location)
        print(salary)
        print('-'*40)
    
    try:
        # If the next symbol is an arrow use the below code
#         next_anchor = soup.find('a', {"aria-label": "Next"})
#         next_page_url = next_anchor['href']
#         browser.click_link_by_partial_href(next_page_url)

        # If the next page is "Next" use this code
        browser.click_link_by_partial_text('Next')


        print("********************Success!!********************")
    except:
        print("********************Did not go to next page!!********************")
        break
```
